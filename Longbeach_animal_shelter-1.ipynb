```python
import requests
import pandas as pd

# Define API URL
dataset_id = "animal-shelter-intakes-and-outcomes"
base_url = f"https://data.longbeach.gov/api/explore/v2.1/catalog/datasets/{dataset_id}/records"

# Initialize empty list to store all results
all_records = []
limit = 100
offset = 0  # Start at the first record

while True:
    # Make API request with pagination
    params = {"limit": limit, "offset": offset}
    response = requests.get(base_url, params=params)

    if response.status_code == 200:
        data = response.json()
        
        # Check if there are results
        if 'results' in data and len(data['results']) > 0:
            all_records.extend(data['results'])  # Add to dataset
            offset += limit  # Move to the next batch
        else:
            break  # Exit loop when no more records remain
    else:
        print(f"Error {response.status_code}: {response.text}")
        break

# Convert to DataFrame and save
if all_records:
    df = pd.DataFrame(all_records)
    df.to_csv("longbeach_full_dataset.csv", index=False, encoding="utf-8")
    print("Data saved to longbeach_full_dataset.csv")
else:
    print("No records found.")
```

```python
import requests
import pandas as pd

# Define API URL
dataset_id = "animal-shelter-intakes-and-outcomes"
base_url = f"https://data.longbeach.gov/api/explore/v2.1/catalog/datasets/{dataset_id}/records"

# Initialize empty list to store results
all_records = []
limit = 100
offset = 0  # Start at the first record
max_records = 10000  # API restriction

while offset < max_records:
    params = {"limit": limit, "offset": offset}
    response = requests.get(base_url, params=params)

    if response.status_code == 200:
        data = response.json()
        
        # Check if results exist
        if 'results' in data and len(data['results']) > 0:
            all_records.extend(data['results'])  # Add batch to dataset
            offset += limit  # Move to the next batch
        else:
            break  # Exit loop when no more records remain
    else:
        print(f"Error {response.status_code}: {response.text}")
        break

# Save data if records were retrieved
if all_records:
    df = pd.DataFrame(all_records)
    df.to_csv("longbeach_limited_dataset.csv", index=False, encoding="utf-8")
    print("Data saved to longbeach_limited_dataset.csv")
else:
    print("No records found.")
```
